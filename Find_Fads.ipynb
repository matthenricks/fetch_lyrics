```python
import pymongo
import numpy
import pandas as pd
import datetime

import csv
from unidecode import unidecode
import re

from gensim import corpora
from nltk.stem.porter import *
from nltk.corpus import stopwords
from nltk.tokenize import WordPunctTokenizer
```

```python
# Set up the infrastructure to clean the text
stopset = set(stopwords.words('english'))
stemmer = PorterStemmer()
def cleanText(column):
    tokens = WordPunctTokenizer().tokenize(column)
    clean = [token.lower() for token in tokens if token.lower() not in stopset and len(token) > 2]
    final = [stemmer.stem(word) for word in clean]
    return final
```

```python
# Import the dictionary
dictionary = corpora.Dictionary.load('reduced_dictionary.mm')
#print dictionary.items()
```

Pull in the mongo data and output it by year

```python
# mongoimport --db lyrics --type json --collection songs --headerline --file lyricScrape.json
# mongoimport --db lyrics --type json --collection errors --headerline --file lyricErrors.json

from pymongo import MongoClient 
client = MongoClient()
db = client.lyrics
songs = db.songs.find({})
db.word_year.drop()
db.create_collection("word_year")

for row in songs:
    column = unidecode(row['lyrics'])
    column = re.sub('  +', ' ', column)
    column = re.sub('\n', ' ', column)
    column = column.strip().strip('"').strip("'").lower().strip()
    if column != "":
        myString = ""
        for x in cleanText(column):
            myString += x + ' '
        
        doc = dictionary.doc2bow(myString.split())
        for wordID, count in doc:
            try:
                if row['year'] != None:
                    db.word_year.insert({"year": row['year'], "wordid": wordID, "count":count})
            except KeyError as e:
                # print row
                print "error"
```

```python
song_stats = db.song_stats.find({})
# print song_stats.count()
song_stats.sort("value.stddev", direction=1)

# Find the variance of the times
variant = {}

i = 0
for x in song_stats:
    # print x
    # print dictionary[x["_id"]], x["value"]["stddev"], x["value"]["avg"], x["value"]["count"]
    pos = round(x["value"]["avg"])
    if (variant.has_key(pos) == False):
        variant[pos] = (x["value"]["stddev"], x["value"]["count"], 1)
    else:
        (s, c, n) = variant[pos]
        variant[pos] = (x["value"]["stddev"] + s, c + x["value"]["count"], n + 1)

```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%pylab inline
```

```python
year = []
stddev = []

for key in variant:
    (s, c, n) = variant[key]
    year += [key]
    stddev += [s / n]


df = pd.DataFrame(stddev, index=year)
df.plot(kind='bar')
```

Attempt to visualize the progression of one specified name notice how the log reduces the recent songs

```python
word_year = db.word_year.find({"wordid":dictionary.token2id["nightshad"]})
word_year.sort("year", direction=-1)
variant = {}

for x in word_year:
    pos = round(float(x["year"]))
    if (variant.has_key(pos) == False):
        variant[pos] = log(x["count"] + 1) #log
    else:
        variant[pos] += log(x["count"] + 1) #log
        
    # print dictionary[x["wordid"]], x
    
year = []
summ = []

for y in variant:
    year += [y]
    summ += [variant[y]]

df = pd.DataFrame(summ, index=year)
df.plot(kind='bar')
```
