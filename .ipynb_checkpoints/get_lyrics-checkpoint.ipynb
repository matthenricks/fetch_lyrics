Song and LyricError Class

```python
# Intialize DataFrames

from json import JSONEncoder 

class Song:
    def __init__(self, artist, track, url, lyrics):
        self.artist = artist
        self.track = track
        self.url = url
        self.lyrics = lyrics
    
    def __str__(self): 
        result = "Artist: " + self.artist
        result += ", Track: " + self.track 
        result += ", URL: " + self.url
        return result 
    
    def __repr__(self): 
        return self.__str__()
    
    def jsonable(self):
        return self.__dict__

class LyricError: 
    def __init__(self, artist, track, errorType, errorMessage):
        self.artist = artist
        self.track = track
        self.errorType = errorType
        self.errorMessage = errorMessage
        
    def __str__(self):
        result = "Artist: " + self.artist
        result += ", Track: " + self.track 
        result += ", ErrorType: " + self.errorType
        result += ", Msg: " + self.errorMessage
        return result 
    
    def __repr__(self): 
        return self.__str__()
    
    def jsonable(self):
        return self.__dict__
    
def ComplexHandler(Obj):
    if hasattr(Obj, 'jsonable'):
        return Obj.jsonable()
    else:
        raise TypeError, 'Object of type %s with value of %s is not JSON serializable' % (type(Obj), repr(Obj))
```

Functions to pull from song API

```python
# lyrics&music API
song_base = "http://api.lyricsnmusic.com/songs?api_key="

# Send out a series of requests for a series of songs across APIs
def findSongAPI(songs, mistakes = 5):
    
    urls = []
    counts = []
    counter = 0
    for song in songs:
        songs[counter]['artist'] = re.sub(r'[^\x00-\x7F]+',' ', song['artist'])
        songs[counter]['track'] = re.sub(r'[^\x00-\x7F]+',' ', song['track'])
        try:
            artist = urllib2.quote(song['artist'].encode("utf-8"))
            track = urllib2.quote(song['track'].encode("utf-8"))
            counts += [counter]
            counter += 1
        except UnicodeDecodeError as e:
            curr_error = LyricError(song['artist'], song['track'], "URL CREATION ERROR", str(e))
            errors.insert(curr_error.__dict__)
            print "URL CREATION ERROR", str(e)
            counter += 1
        urls += [song_base+song['api']+"&artist="+artist+"&track="+track]

    song_requests = (grequests.get(u) for u in urls)
    song_results = grequests.map(song_requests)
    
    counts2 = []
    urls = []
    counter = 0
    for i in range(0, len(song_results)):
        result = song_results[i]
        if result is not None and result.ok:
            try:
                lyric_url = result.json()[0]['url']
                urls += [lyric_url.encode()]
                counts2 += [i]
            except (AttributeError, ValueError, IndexError, ConnectionError, UnicodeDecodeError, UnboundLocalError, Exception) as e:
                data = songs[counts[i]]
                curr_error = LyricError(data['artist'], data['track'], "API Call Error", str(e))
                errors.insert(curr_error.__dict__)
                print "API Call Error", str(e)
        else:
            # Handle bad song
            data = songs[counts[i]]
            curr_error = LyricError(data['artist'], data['track'], "API Call Error", "Result Not Okay")
            errors.insert(curr_error.__dict__)
            print "API CALL ERROR"
            #print result
        result.close()
    
    song_requests = (grequests.get(u) for u in urls)
    song_results = grequests.map(song_requests)
    
    counter = 0
    lyric_array = []
    for i in range(0, len(song_results)):
        result = song_results[i]
        #print result
        if result is None or not result.ok:
            data = songs[counts[counts2[i]]]
            curr_error = LyricError(data['artist'], data['track'], "Song URL Result Error", str(urls[i]))
            errors.insert(curr_error.__dict__)
            print "SONG URL RESULT ERROR"
            result.close()
            continue
            
        soup = BeautifulSoup(result.text)
        soup = soup.find('pre', attrs={'itemprop':'description'})
        
        if soup is None:
            data = songs[counts[counts2[i]]]
            curr_error = LyricError(data['artist'], data['track'], "URL Traversal Error", str(urls[i]))
            errors.insert(curr_error.__dict__)
            print "URL Traversal Error"
            result.close()
            continue
        
        lyrics = soup.text.encode('utf-8')
        lyrics = lyrics.replace("\r\n", " ")
        lyric_array += [(lyrics, counts[counts2[i]], i)]
        result.close()
        
    
    results = []
    for (lyric, count, urlID) in lyric_array:
        s = songs[count]
        results += [Song(s['artist'], s['track'], urls[urlID], lyric)]
        
    return results
```
